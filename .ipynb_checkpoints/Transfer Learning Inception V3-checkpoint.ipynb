## Transfer Learning Inception V3 using Keras

```python
from tensorflow.compat.v1 import ConfigProto
from tensorflow.compat.v1 import InteractiveSession

config = ConfigProto()
config.gpu_options.per_process_gpu_memory_fraction = 0.5
config.gpu_options.allow_growth = True
session = InteractiveSession(config=config)
```

```python
# import the libraries as shown below

from tensorflow.keras.layers import Input, Lambda, Dense, Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras.applications.inception_v3 import preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img
from tensorflow.keras.models import Sequential
import numpy as np
from glob import glob
```

```python
# re-size all the images to this
IMAGE_SIZE = [256, 256]

train_path = '/home/ociadmin/Documents/Shobnom/DLFirst/Tomato-leaf-disease-classification/Dataset/train'
valid_path = '/home/ociadmin/Documents/Shobnom/DLFirst/Tomato-leaf-disease-classification/Dataset/valid'
```

```python
# Import the InceptionV3 library as shown below
# Here we will be using imagenet weights

inception = InceptionV3(input_shape=IMAGE_SIZE + [3], weights='imagenet', include_top=False)
```

```python
# don't train existing weights
for layer in inception.layers:
    layer.trainable = False
```

```python
# useful for getting number of output classes
folders = glob('/home/ociadmin/Documents/Shobnom/DLFirst/Tomato-leaf-disease-classification/Dataset/train/*')
```

```python
len(folders)
```

```python
# the layers of this model
x = Flatten()(inception.output)
```

```python
prediction = Dense(len(folders), activation='softmax')(x)

# create a model object
model = Model(inputs=inception.input, outputs=prediction)
```

```python

# view the structure of the model
model.summary()
```

```python
# tell the model what cost and optimization method to use
model.compile(
  loss='categorical_crossentropy',
  optimizer='adam',
  metrics=['accuracy']
)
```

```python
# Use the Image Data Generator to import the images from the dataset
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)

test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
# Make sure you provide the same target size as initialied for the image size
training_set = train_datagen.flow_from_directory('/home/ociadmin/Documents/Shobnom/DLFirst/Tomato-Leaf-Disease-Prediction/Datasets/train',
                                                 target_size = (224, 224),
                                                 batch_size = 32,
                                                 class_mode = 'categorical')
```

```python
test_set = test_datagen.flow_from_directory('/home/ociadmin/Documents/Shobnom/DLFirst/Tomato-Leaf-Disease-Prediction/Datasets/test',
                                            target_size = (224, 224),
                                            batch_size = 32,
                                            class_mode = 'categorical')
```

```python
# fit the model

r = model.fit_generator(
  training_set,
  validation_data=test_set,
  epochs=10,
  steps_per_epoch=len(training_set),
  validation_steps=len(test_set)
)
```

```python
import matplotlib.pyplot as plt
```

```python
# plot the loss
plt.plot(r.history['loss'], label='train loss')
plt.plot(r.history['val_loss'], label='val loss')
plt.legend()
plt.show()
plt.savefig('LossVal_loss')

# plot the accuracy
plt.plot(r.history['accuracy'], label='train acc')
plt.plot(r.history['val_accuracy'], label='val acc')
plt.legend()
plt.show()
plt.savefig('AccVal_acc')
```
